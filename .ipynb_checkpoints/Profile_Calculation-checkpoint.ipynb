# Profile Calculation

In this notebook, we are going to calculate the profiles of the trench/volcano pairs and analyze the results.

```python
# Importing the modules/libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
import shapely.geometry as geo
from scipy.interpolate import griddata
from scipy.interpolate import interp1d
import geopy.distance
```

```python
# For our calculations, we need the data created in the last notebook. 
# The shapefile for South America is used for plotting

data = pd.read_pickle("./data/trenchvolcanoes")

america = gpd.read_file("./data/Continent/SouthAmerica.shp")

# Loading the depth grid from the slab2 database

fp = "./data/slab2/sam_slab2_dep_02.23.18.xyz"

data2 = np.loadtxt(fp, delimiter = ",")

dataset = pd.DataFrame({'Longitude':data2[:,0]-360,
                        'Latitude':data2[:,1], 
                        "Depth":data2[:,2]})
```

```python
# Masking the values where Depth = NaN

df = dataset[np.isfinite(dataset['Depth'])]
```

```python
# Creating coordinate and depth arrays

points_depth = df[["Longitude", "Latitude"]].values

values_depth = df[["Depth"]].values
```

```python
# Setting up dictionaries where we will store the different lists of values (could also be lists)

dctx = {} # is saving the Longitudes of the points in a list for each pair
dcty = {} # saving the Latitudes of the points in a list for each pair
dctc = {} # is saving the Longitudes + Latitudes as a coordinate pair
dctp = {} # is saving the depths of the points 
dctd = {} # is saving the distances of the points from the trench
```

```python
# This for loop is going through each pair of a point on the trench and a volcano and is 
# creating 50 points between them for which it is calculating the depth


for i, (x1, y1, x2, y2)  in enumerate(zip(data["Longitude"], data["Latitude"], data["lon_vol"], data["lat_vol"])):
    dctx['%s' % i] = []
    dcty['%s' % i] = []
    dctc['%s' % i] = []
    dctp['%s' % i] = []
    x = np.linspace(x1,x2) # creating an array of equally distributed values between the longitudes
    dctx[i] = x # saving the values
    m = (y2-y1)/(x2-x1) # creating a slope 
    b = y1 - m * x1 
    dcty[i] = m *(x) + b # the latitudes are calculated with a linear function of the longitudes
    dctc[i] = np.column_stack((x, dcty[i])) # saving the coordinate pairs
    dctp[i] = griddata(points_depth, values_depth, dctc[i]) # calculating their respective depths
    coords_t = (x1, y1)
    coords_vol = (x2, y2)
    distance = geopy.distance.distance(coords_t, coords_vol).km # distance between the coordinates
    dctd[i] = np.linspace(0,distance) # array of distance in km between point and trench start
```

```python
# plotting our results

for i in range(len(data)):
    plt.plot(dctd[i], dctp[i])
    
plt.gca().set_aspect('equal', adjustable='box')
```

```python
# simple pythagorean theorem

def dist(x,y):   
    dist = x**2+y**2
    return np.sqrt(dist)
```

```python
dctl = {} # length between
lengths = [] # saving the slab lengths of each trench-volcano pair
ymax = [] # saving the depth of the slab at the volcano
dcta = {} # saving the slab dip 

for i in range(len(dctd)):
    dctl['%s' % i] = []
    ymax.append(np.amax(abs(dctp[i])))
    x1 = 0
    y1 = 0
    dcta[i] = np.rad2deg(np.arcsin(np.amax(abs(dctp[i]))/dist(np.amax(dctd[i]),np.amax(abs(dctp[i])))))
    for p, q in zip(dctd[i], dctp[i]):
        dctl['%s' % i].append(dist(p-x1,q-y1))
        x1 = p
        y1 = q
    lengths.append(sum(dctl['%s' % i]))
```

```python
lengths =  [arr.tolist() for arr in lengths]
```

```python
lengths = [item for sublist in lengths for item in sublist]
```

Let's see how the lengths of the profile differ to using a normal hypothenuse:

```python
percentages = [] # saving the differences as percentages in the list

for i, length in enumerate(lengths):
    a = np.max(abs(dctp[i]))
    b = np.max(abs(dctd[i]))
    c = np.sqrt(a**2 + b**2)
    difference = length - c
    percentage = difference / c * 100
    percentages.append(percentage)
```

```python
percentages = np.asarray(percentages)
```

```python
percentages = percentages[~np.isnan(percentages)]
```

```python
np.mean(percentages)
```

```python
# Now we can plot the differences with the indices 

fig, ax = plt.subplots(figsize=(16,12))
plt.plot(percentages, linewidth = 3)
plt.suptitle("Increase by Using Approximated Length Instead of Hypothenuse ", size = 28)
plt.xlim(0,161)
plt.ylabel("Increase in length [%]",size = 24)
plt.xlabel("Index of the profiles",size = 24)
plt.grid()
```

We can see that there is a small difference in length between both methods with a mean of ~2.6 percent. 

# McCaffrey Equation

Now that we have the length of the profiles, we can use the modified McCaffrey equation by Whipp et al. (2014) for the estimation of strain partitioning. 


$$sin(γ)_{crit} = {z_{slab} \over Lk_{t}}{sin(\phi)_{s}\over sin(\phi)_{t}}$$

Where $z_{slab}$ is the depth of the slab at the location of the volcano. $L$ is the length of the contact area between the two plates. $sin(\phi)_{s}$ is the internal angle of friction for the continental strike-slip fault and $sin(\phi)_{t}$ is the internal angle of friction of the subduction interface and $k_{t}$ is defined as followed:

$$k_{t} = {1 \over 1-2sin(\phi)_{t}\sqrt{3}}$$ 

$k_{t}$ is used to estimate compression or extension according to Rankine (1857).



Below we have the function that is calculating the critical obliquity $sin(γ)_{crit}$:

```python
def critical_obliquity(z_slab,phi_s, phi_t, length):
    '''Calculating the critical obliquity of individual points on the trench 
    with a modified version of the McCaffrey paper.'''
    k = 1 / (1-2*(np.sin(np.deg2rad(phi_t))/np.sqrt(3))) # calculating kt
    critical_obliquity_sin = (z_slab/(length*k))*(np.sin(np.deg2rad(phi_s))/np.sin(np.deg2rad(phi_t)))
    x = np.rad2deg(np.arcsin(critical_obliquity_sin)) # convert radians to degrees 
    critical_obliquity = round(x,2)
    return critical_obliquity
```

Plugging in some possible values:

```python
critical_obliquity(z_slab = 120, phi_s = 15, phi_t = 8, length = 300)
```

It's working!

```python
# Taking the highest depths from all the profiles

depthspoints = []
for i in range(len(data)):
    depthspoints.append(np.amax(abs(dctp[i])))
```

```python
# Adding it to our dataframe

data["Lengths"] = lengths
data["Depths"] = depthspoints
```

The volcanic regions of the Andes are divided into three different zones:
 * Northern Volcanic Zone
 * Central Volcanic Zone
 * Southern Volcanic Zone

```python
# Dividing the zones by using the Latitudes

northern = data[data["Latitude"]>-10]
central = data[(data["Latitude"]<-10) & (data["Latitude"]>-30)]
southern = data[data["Latitude"]<-30]
```

```python
# Creating GeoDataFrames for every zone

gdf_northern = gpd.GeoDataFrame(northern, geometry = [geo.Point(x,y) for x,y in zip(northern.lon_vol, northern.lat_vol)])
gdf_central = gpd.GeoDataFrame(central, geometry = [geo.Point(x,y) for x,y in zip(central.lon_vol, central.lat_vol)])
gdf_southern = gpd.GeoDataFrame(southern, geometry = [geo.Point(x,y) for x,y in zip(southern.lon_vol, southern.lat_vol)])
```

```python
# Calculating the critical obliquities for different angles of friction

critical_obliquities_11_7_5 = []
critical_obliquities_11_4 = []
critical_obliquities_5_2 = []

for length, depth in zip(data["Lengths"],data["Depths"]):
    critical_obliquities_11_7_5.append(critical_obliquity(z_slab = depth, phi_s = 11, phi_t = 7.5, length = length))
    critical_obliquities_11_4.append(critical_obliquity(z_slab = depth, phi_s = 11, phi_t = 4, length = length))
    critical_obliquities_5_2.append(critical_obliquity(z_slab = depth, phi_s = 5, phi_t = 2, length = length))
```

```python
# Calculating the differences between the obliquity angles and the possible critical obliquities

differences = data["Obliquity"] - critical_obliquities_11_7_5
```

```python
data["Differences"] = differences
```

```python
data = data[(data["Differences"]<50) & (data["Differences"]>0)]
```

```python
gdf = gpd.GeoDataFrame(data, geometry = [geo.Point(x,y) for x,y in zip(data.Longitude, data.Latitude)])
```

```python
# removing the NaNs

cleaned_lengths = [x for x in lengths if ~np.isnan(x)]

cleaned_depths = [x for x in depthspoints if ~np.isnan(x)] 
```

```python
# creating means of the lengths and depths so we can plug them in our function

cl = np.mean(cleaned_lengths)

cp = np.mean(cleaned_depths)
```

```python
# creating critical obliquities for fixed phi_s values

val1 = np.linspace(0,10,50)
phi_s_15 = []
phi_s_10 = []
phi_s_75 = []
phi_s_5 = []
for v in val1:
    phi_s_15.append(critical_obliquity(z_slab = cp, phi_s = 15, phi_t = v, length = cl))
    phi_s_10.append(critical_obliquity(z_slab = cp, phi_s = 10, phi_t = v, length = cl))
    phi_s_75.append(critical_obliquity(z_slab = cp, phi_s = 7.5, phi_t = v, length = cl))
    phi_s_5.append(critical_obliquity(z_slab = cp, phi_s = 5, phi_t = v, length = cl))
```

```python
# plotting our results

phi_s_lists = [phi_s_15, phi_s_10, phi_s_75, phi_s_5]
numbers = [15, 10, 7.5, 5]

fig, ax = plt.subplots(figsize=(9,6))

for i in range(len(phi_s_lists)):
    
    plt.plot(val1, phi_s_lists[i], label= '\u03C6\u209b = %s ' % numbers[i])

plt.legend(loc=1, prop={'size': 18})
plt.xlim(1.6,10)
plt.suptitle("Relation Between Critical Obliquities and \u03C6\u209c Values for Fixed \u03C6\u209b ", 
             size = 20)

plt.xlabel('\u03C6\u209c', size=24)
plt.ylabel("$\u03B3_{crit}$",size=24)

plt.savefig("phi_s")
```

```python
# same with fixed phi_t values

val = np.linspace(5,15,50)
phi_t_8 = []
phi_t_6 = []
phi_t_4 = []
phi_t_2 = []
for v in val:
    phi_t_8.append(critical_obliquity(z_slab = cp, phi_s = v, phi_t = 8, length = cl))
    phi_t_6.append(critical_obliquity(z_slab = cp, phi_s = v, phi_t = 6, length = cl))
    phi_t_4.append(critical_obliquity(z_slab = cp, phi_s = v, phi_t = 4, length = cl))
    phi_t_2.append(critical_obliquity(z_slab = cp, phi_s = v, phi_t = 2, length = cl))
```

```python
# plotting these as well

phi_t_lists = [phi_t_8, phi_t_6, phi_t_4, phi_t_2]
numberst = [8, 6, 4, 2]

fig, ax = plt.subplots(figsize=(9,6))

for i in range(len(phi_t_lists)):
    
    plt.plot(val, phi_t_lists[i], label= '\u03C6\u209c = %s ' % numberst[i])

plt.legend(loc=1, prop={'size': 18})
plt.xlabel('\u03C6\u209b', size=24)
plt.ylabel("$\u03B3_{crit}$",size=24)
plt.suptitle("Relation Between Critical Obliquities and \u03C6\u209b Values for Fixed \u03C6\u209c", 
             size = 20)
plt.xlim(4,15)
plt.savefig("phi_t")
```
